```python
import pandas as pd
from numpy import nan
from os import listdir
from os.path import isfile, join
```

```python
data = pd.read_excel('Предложения по отзыву (ВР 211,214) - с 21.09.2022.xlsx')
data.info()
```

```python
data['Наименование ТОВД'] = data['Наименование ТОВД'].fillna(method='ffill')
data['КБК'] = data['КБК'].fillna(method='ffill')
data = data.loc[data['КБК'] == '0302 0830292019 211']
data
```

```python
data.info()
```

```python
data1 = data.loc[:, ['Наименование ТОВД', 'КБК', 'Наименование задания ГОЗ','Номенклатурная позиция', 'На сумму, руб.']]
data2 = data1.pivot_table(index=['Наименование ТОВД', 'КБК'], columns = ['Наименование задания ГОЗ', 'Номенклатурная позиция'], values=['На сумму, руб.'], aggfunc='sum', sort=False)
#data2 *= -1
data2
```

```python
data['Наименование задания ГОЗ'].sort_values().unique()
```

```python
with pd.ExcelWriter('Отзывы.xlsx') as writer:
    data2.to_excel(writer)
```
